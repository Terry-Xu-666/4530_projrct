```python
def add_0(df):
    ticker_code=df[['代码','名称']]
    for i in range(ticker_code.shape[0]):
        if len(str(ticker_code.loc[df.index[i],'代码']))<6:
            add_code="0"*(6-len(str(ticker_code.loc[df.index[i],'代码'])))
            ticker_code.loc[df.index[i],'代码']=add_code+str(ticker_code.loc[df.index[i],'代码'])
    return ticker_code
        
    
```

```python
import time
def outstanding(selected,n):
    for i in tqdm(range(n,len(selected))):
        try:
            code=selected.loc[selected.index[i],'代码']
            stock_individual_info_em_df = ak.stock_individual_info_em(symbol=str(code))
            slected_row=stock_individual_info_em_df.iloc[7,1]

            selected.loc[selected.index[i],'outstandings']=slected_row
        except:
            print(f"错误位置{i}")
            return selected
    return selected

new_selected=outstanding(new_selected,0)
```

```python
import yfinance as yf
from tqdm import tqdm
```

```python
stocks=[]
from datetime import datetime,timedelta
for i in tqdm(ticker):
    try:
        stock=yf.Ticker(i)
        listing_date=stock.history(period='max').index[0].to_pydatetime().date()
        twenty_years=(datetime.today()-timedelta(days=20*365)).date()
        if listing_date<twenty_years:
            stocks.append(i)
    except:
        print(f"stock{i}have something wrong")
        continue
    
```

```python
ticker_20=pd.DataFrame(stocks)
ticker_20.to_csv('ticker_20.csv')
```
