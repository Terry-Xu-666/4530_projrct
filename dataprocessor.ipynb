```python
data2=pd.read_csv('ticker.csv')
ticker=data2['Ticker']
ticker
```

```python
import yfinance as yf
from tqdm import tqdm
```

```python
stocks=[]
from datetime import datetime,timedelta
for i in tqdm(ticker):
    try:
        stock=yf.Ticker(i)
        listing_date=stock.history(period='max').index[0].to_pydatetime().date()
        twenty_years=(datetime.today()-timedelta(days=20*365)).date()
        if listing_date<twenty_years:
            stocks.append(i)
    except:
        print(f"stock{i}have something wrong")
        continue
    
```

```python
ticker_20=pd.DataFrame(stocks)
ticker_20.to_csv('ticker_20.csv')
```
