```python
def z_score(df):
    df=df.loc[:,['Anual_ret','std','calmar_ratio','sharp_ratio']]
    df['std']=-1*df['std']
    z=(df-df.mean(axis=0))/df.std(axis=0)
    z=z.sum(axis=1)
    df.loc[:,'z_score']=z/4
    return df
z_score(df.sort_values(by='z_score',ascending=False))    
```
